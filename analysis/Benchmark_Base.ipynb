## Benchmark 1

```python
folder_path = '../credentials/results/Benchmark1/'
dfb1 = load_data(folder_path)
```

```python
dfb1F = dfb1[dfb1['Henrique'] == False]
dfb1T = dfb1[dfb1['Henrique'] == True]
```

```python
dataH = [0.156, 0.210, 0.472]
```

```python
def plot_results_box_plot(dfs, legend):
    
    fig, ax = plt.subplots()

    # Creating axes instance
    plt.boxplot(dfs)
    ax.legend(legend, fontsize=12)
 
    # show plot
    plt.show()

plot_results_box_plot([dfb1T['RTT'].tolist(), dataH],
                      legend=['Com match de conteúdo', 'Sem match de conteúdo', 'Sem match de conteúdo - Artigo'])
```
